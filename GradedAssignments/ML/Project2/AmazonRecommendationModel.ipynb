```python
import pandas as pd

amzMovieTvRatings = pd.read_csv("Amazon - Movies and TV Ratings.csv", low_memory=False)

amzMovieTvRatings.head()
```

```python
# Transpose the dataframe to be able to calculate the total votes and votes average
amzMovieTvRatingsT = amzMovieTvRatings.T
amzMovieTvRatingsT.head()
```

```python
vote_average = [0,]
vote_count = [0,]

#range starts from 1 as the first row is user Id and not the movie name.
for indx in range(1, len(amzMovieTvRatingsT)) :
    vote_average.append(amzMovieTvRatingsT.iloc[indx].mean(skipna=True, axis=0))
    vote_count.append((4848 - amzMovieTvRatingsT.iloc[indx].isna().sum()))

#Answer to 2) What is the average rating for each movie?    
amzMovieTvRatingsT["vote_average"] = vote_average
amzMovieTvRatingsT["vote_count"] = vote_count
amzMovieTvRatingsT.head()
```

```python
#Calculate of the mean of vote average
C = amzMovieTvRatingsT["vote_average"].mean()

print("Value of C is ", C)
```

```python
#Calculate the minimum number of votes required to be in the chart, m

m = amzMovieTvRatingsT["vote_count"].quantile(0.75)
print("value of m is ",m)
```

```python
#filter out all the qualifying movies into anothe DF
filteredAmzMoviesTvRatings = amzMovieTvRatingsT.copy().loc[amzMovieTvRatingsT["vote_count"] >= m]
filteredAmzMoviesTvRatings.shape
```

```python
amzMovieTvRatingsT.shape
```

```python
# Customized function to calculate weighted rating of each movie. 
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
#weighted_rating column to be abbed newly
amzMovieTvRatingsT["weighted_rating"] = amzMovieTvRatingsT.apply(weighted_rating, axis=1)
filteredAmzMoviesTvRatings["weighted_rating"] = filteredAmzMoviesTvRatings.apply(weighted_rating, axis=1)
filteredAmzMoviesTvRatings.head()
```

```python
#Answer to 1) Which movies have maximum views/ratings?
#Answer to 2)Define the top 5 movies with the maximum ratings.
sortedRatingsOfAmz = filteredAmzMoviesTvRatings.sort_values("weighted_rating", ascending=False)
sortedRatingsOfAmz.head()
```

```python
#Answer to 3)Define the top 5 movies with the least audience
sortedVoteCountOfAmz = sortedRatingsOfAmz.sort_values("vote_count", ascending=True)
sortedVoteCountOfAmz.head()
```

```python
# Build a machine learning recommendation algorithm which provides the ratings for each of the users.
#Collaborative recommender system
# Recommender 1) Divide the data into training and test data
# Recommender 2) Build a recommendation model on training data
# Recommender 3) Make predictions on the test data

#For surprise package, in anaconda command promt execute > conda install -c conda-forge scikit-surprise

from surprise import Dataset
from surprise import Reader
```

```python
amzRatings = pd.read_csv("Amazon - Movies and TV Ratings.csv", low_memory=False)

#Convert the given data into a DF with 3 columns - User Id, Movie Id and rating that has been given to "a" movie Id, by the user Id

#Since userId is alphanumeric, am mainitaing a list of user Id in numberic to alphanumeric mapping. 
#Implies userIdIndx2Val[1] will return to me the UserId "A3R5OBKS7OM2IR" (from original data set) of an user with numeric Id = 1
#We have totally 4848 users and 206 movies.

userIdIndx2Val = ["0",]
userIds = list()
movieIds = list()
ratings = list()

for indx in range(1,len(amzRatings)):
    amzRow = list(amzRatings.loc[indx-1])
    userIdIndx2Val.append(amzRow[0])
    for innerIndx in range(1, len(amzRow)) :
        if (not pd.isna(amzRow[innerIndx])) :
            userIds.append(indx)
            movieIds.append(innerIndx)
            ratings.append(amzRow[innerIndx])

data = {
    "userId" : userIds,
    "movieId" : movieIds,
    "rating" : ratings
}

amzUserMovieRating = pd.DataFrame(data)
amzUserMovieRating.head()
#Yay! data frame is now ready. We can start with the collaborative recommender system
```

```python
from surprise import SVD
from surprise import Dataset
from surprise import accuracy
from surprise.model_selection import train_test_split
```

```python
reader = Reader()
data = Dataset.load_from_df(amzUserMovieRating[['userId', 'movieId', 'rating']], reader)

# Step 1 - Divide the data into training and test data
# test set is made of 25% of the ratings.
trainset, testset = train_test_split(data, test_size=.25)
```

```python
svdAlgorithm = SVD()

#Step 2 - Build a recommendation model on training data
svdAlgorithm.fit(trainset)
```

```python
# Step 3 - Make predictions on the test data
predictions = svdAlgorithm.test(testset)

# Then compute RMSE
accuracy.rmse(predictions)
```

```python
#Predicting for userId 1 for movie Id 2
svdAlgorithm.predict(1,2, 5.0, verbose=True)
```
